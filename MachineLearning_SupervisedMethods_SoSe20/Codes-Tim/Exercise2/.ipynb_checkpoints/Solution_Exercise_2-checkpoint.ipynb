```python
import numpy as np
import random 
import matplotlib.pyplot as plt
import pandas as pd

def loadCSV(filename):
    f = open(filename)
    data = np.loadtxt(f, delimiter=',')
    X = data[:, 1:]
    y = data[:, 0]
    return X, y

X, y = loadCSV("mushroom.csv")
```

## We plot the banana data

```python
print("The shape of X is " + str(X.shape))
print("The shape of y is " + str(y.shape))
```

```python
def column(matrix, i):
    return [row[i] for row in matrix]
```

```python
def plot_separated(data, labels):
    conc_array=np.concatenate((data[:,0], data[:,1],labels)).reshape(3,len(data))
    true_array = [row for row in np.transpose(conc_array) if row[2]==1]
    false_array = [row for row in np.transpose(conc_array) if row[2]==-1]
    plt.scatter(column(true_array,0),column(true_array,1), c='b', marker='x', label='1')
    plt.scatter(column(false_array,0), column(false_array,1), c='r', marker='d', label='-1')
    plt.legend(loc='upper left')
    plt.show()
```

```python
plot_separated(X,y)
print("The above graphic displays the banana dataset.")
```

## Generate a set of 10 hypotheses

```python
np.random.seed(0)

X_training=X[0:int(len(y)/2),:]
X_test=X[int((len(y))/2):,:]
y_training=y[0:int(len(y)/2)]
y_test=y[int((len(y))/2):]


h=np.zeros((1000,len(X[0,:])))
for i in range(0,1000):
    h[i]=np.random.normal(loc=0.0, scale=1, size=len(X[0,:]))
```

```python
error=np.zeros(1000)
for i in range(0,1000):
    print(sum(abs(np.sign(np.dot(X_training,h[i]))-y_training))/(2*len(y_training)))
    error[i]=sum(abs(np.sign(np.dot(X_training,h[i]))-y_training))/(2*len(y_training))
best_hypot=np.argmin(error)
```

```python
print("The test error is " + str(sum(abs(np.sign(np.dot(X_test,h[best_hypot]))-y_test))/(2*len(y_test))))
print("The training error is " + str(sum(abs(np.sign(np.dot(X_training,h[best_hypot]))-y_test))/(2*len(y_training))))
```
